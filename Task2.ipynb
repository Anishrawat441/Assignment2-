# Task 2: Process Multiple Orders(for loop)


```python
# 1. Given a list of order amounts: orders=[1200,2500,800,1750,3000] use a for loop to apply the discount rules 
# from task 1 to each order and print a summary table
# showing: order_amount-> discount%->final_amount.

orders=[1200, 2500, 800, 1750, 3000]

for order_amount in orders:
    if order_amount >= 2000:
        discount_percent = 15
    elif 1500 <= order_amount < 2000:
        discount_percent = 10
    elif 1000 <= order_amount < 1500:
        discount_percent = 7
    else:
        discount_percent = 0

    discount = (discount_percent / 100) * order_amount
    final_amount = order_amount - discount

    print(f"{order_amount} -> {discount_percent}% -> {final_amount}")
```

```python

orders=[1200, 2500, 800, 1750, 3000]
total_revenue=0
for order_amount in orders:
    if order_amount >= 2000:
        discount_percent = 15
    elif 1500 <= order_amount < 2000:
        discount_percent = 10
    elif 1000 <= order_amount < 1500:
        discount_percent = 7
    else:
        discount_percent = 0

    discount = (discount_percent / 100) * order_amount
    final_amount = order_amount - discount
    total_revenue+=final_amount

    print(f"{order_amount} -> {discount_percent}% -> {final_amount}")
    print("The total revenue after discount:",total_revenue)
```
