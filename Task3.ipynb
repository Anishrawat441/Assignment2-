# Task 3: User Menu(while loop + break/continue)

```python
# 1.implement a simple while-loop driven menu that keeps asking the user to 
# choose an action until they press q to quit. menu options: 1--add order amount
#  to a running list(use input()) 2- show all orders and total after applying discounts q- Quit
orders = []

while True:
    print("\nMenu")
    print("1 - Add order amount")
    print("2 - Show all orders and total after discounts")
    print("q - Quit")

    choice = input("Enter your choice: ").lower()

    if choice == "1":
        order_amount = int(input("Enter order amount: "))
        orders.append(order_amount)
        print("Order added.")

    elif choice == "2":
        if not orders:
            print("No orders available.")
        else:
            total_revenue = 0
            print("\nOrder Amount -> Discount % -> Final Amount")
            print("--------------------------------------------")

            for amount in orders:
                if amount >= 2000:
                    discount = 15
                elif 1500 <= amount < 2000:
                    discount = 10
                elif 1000 <= amount < 1500:
                    discount = 7
                else:
                    discount = 0

                final_amount = amount - (amount * discount / 100)
                total_revenue += final_amount

                print(f"{amount} -> {discount}% -> {final_amount}")

            print("--------------------------------------------")
            print("Total revenue after discounts:", total_revenue)

    elif choice == "q":
        print("Program exited.")
        break

    else:
        print("Invalid option. Try again.")
```

```python
# 2. use continue to re show the menu after invalid input and break to exit on q
orders = []

while True:
    print("\nMenu")
    print("1 - Add order amount")
    print("2 - Show all orders and total after discounts")
    print("q - Quit")

    choice = input("Enter your choice: ").lower()

    if choice == "q":
        print("Exiting program.")
        break   # exit the loop

    if choice != "1" and choice != "2":
        print("Invalid choice. Please try again.")
        continue   # re-show the menu

    if choice == "1":
        order_amount = int(input("Enter order amount: "))
        orders.append(order_amount)
        print("Order added.")

    elif choice == "2":
        if not orders:
            print("No orders available.")
            continue

        total_revenue = 0
        print("\nOrder Amount -> Discount % -> Final Amount")
        print("--------------------------------------------")

        for amount in orders:
            if amount >= 2000:
                discount = 15
            elif 1500 <= amount < 2000:
                discount = 10
            elif 1000 <= amount < 1500:
                discount = 7
            else:
                discount = 0

            final_amount = amount - (amount * discount / 100)
            total_revenue += final_amount

            print(f"{amount} -> {discount}% -> {final_amount}")

        print("--------------------------------------------")
        print("Total revenue after discounts:", total_revenue)
```
